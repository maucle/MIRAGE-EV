Python script to control the periscope and record motor states in .csv

```python
import serial
import csv
import time

# Change this to your Arduino port
# PORT = "COM3"          # Windows example
PORT = "/dev/ttyACM0"  # Linux example
# PORT = "/dev/cu.usbmodemXXXX"  # macOS example

BAUD = 115200
CSV_FILE = "angle_log.csv"

speed = float(input("Enter speed in rev/s: "))
duration = float(input("Enter duration in seconds: "))

with serial.Serial(PORT, BAUD, timeout=0.05) as arduino:
    time.sleep(2)  # Arduino resets after serial connection opens

    arduino.reset_input_buffer()

    arduino.write(f"speed {speed}\n".encode())
    time.sleep(0.1)

    arduino.write(b"zero\n")
    time.sleep(0.1)

    arduino.write(b"start\n")
    print("Motor started")

    start_time = time.time()
    stop_sent = False

    with open(CSV_FILE, "w", newline="") as f:
        writer = csv.writer(f)
        writer.writerow(["time_s", "angle_deg"])

        while True:
            now = time.time()
            elapsed = now - start_time

            if elapsed >= duration and not stop_sent:
                arduino.write(b"stop\n")
                stop_sent = True
                print("Stop sent")

            line = arduino.readline().decode(errors="ignore").strip()

            if line.startswith("A "):
                angle = float(line.split()[1])
                writer.writerow([elapsed, angle])
                print(f"{elapsed:.4f}s, {angle:.2f} deg")

            if stop_sent and elapsed >= duration + 0.5:
                break

print(f"Saved CSV to {CSV_FILE}")
```
